### Preparation of Combined Leagues OFFSIDES csv for modelling' 
#### Clean, feature selection, add expanding means, save

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None  # stops the slice default='warn' appearing

%matplotlib inline
```

```python
df1 = pd.read_csv(r"C:\Users\MikeD\Documents\Betting\Football-Data-Combined\FS_Leagues_Data_Combined.csv", index_col=0)         
     
```

```python
df1.head()
```

```python
for col in df1:
    print(col)
```

```python
# cols required for offsides 
df2 = df1[['competition_id',
           'date_unix',
           'home_name',
           'away_name', 
           'season',
           'odds_ft_1',
           'odds_ft_x',
           'odds_ft_2',
           'odds_ft_over25',
           'team_a_offsides',
           'team_b_offsides']]
    
```

```python
df3 = df2.loc[df2['team_a_offsides'] != -1]
```

```python
df3.shape
```

```python
df3.duplicated().sum()
```

```python
df4 = df3.drop_duplicates()
```

```python
df4.duplicated().sum()
```

```python
df4.competition_id.unique()
```

```python
competition_counts = df4['competition_id'].value_counts()

for competition_id, count in competition_counts.items():
    print(f"Competition ID: {competition_id}, Count: {count}")
```

```python
# remove competitions with less than 100 matches

# Get competition_ids where the count is less than 100
filtered_competition_ids = competition_counts[competition_counts < 100].index

# Remove rows with competition_ids that have count less than 100
df4 = df4[~df4['competition_id'].isin(filtered_competition_ids)]
```

```python
df4.shape
```

```python
df4['HW_%'] = round((1/df4['odds_ft_1']) / (1/df4['odds_ft_1'] + 1/df4['odds_ft_x'] + 1/df4['odds_ft_2']),2)  # Add a column giving the Home Win % (HW_%)
df4['X_%'] = round((1/df4['odds_ft_x']) / (1/df4['odds_ft_1'] + 1/df4['odds_ft_x'] + 1/df4['odds_ft_2']),2) # Draw %
df4['AW_%'] = round((1/df4['odds_ft_2']) / (1/df4['odds_ft_1'] + 1/df4['odds_ft_x'] + 1/df4['odds_ft_2']),2) # Away %
```

```python
df4['odds_ft_under25'] = 1/(1.07 - 1/df4['odds_ft_over25'])
```

```python
df4['O_2.5_%'] = round(1/df4['odds_ft_over25'] / (1/df4['odds_ft_over25'] + 1/df4['odds_ft_under25']),2)
df4['U_2.5_%'] = round(1/df4['odds_ft_under25'] / (1/df4['odds_ft_over25'] + 1/df4['odds_ft_under25']),2)
```

```python
df4.drop(['odds_ft_1', 'odds_ft_x', 'odds_ft_2', 'odds_ft_over25', 'odds_ft_under25'], axis = 1, inplace=True)
```

```python
df4.head()
```

```python
# add date column
from datetime import datetime

# Convert Unix timestamps to datetime objects
df4['date_unix'] = pd.to_datetime(df4['date_unix'], unit='s')

# Convert datetime objects to the desired date format (dd/mm/yyyy)
df4['date'] = df4['date_unix'].dt.strftime('%d/%m/%Y')
df4.drop(['date_unix'], axis=1, inplace=True)
df4.head()
```

```python
# Overlay Sup and goal exp
```

```python
dfOU = pd.read_csv('C:/Users/MikeD/Documents/Betting/OverUnder_Exp.csv', index_col=False)
dfOU
```

```python
dfOU = dfOU.drop(['Exp1','Under','Over', 'Un2.5_%'], axis = 1)
dfOU.drop_duplicates(subset='Ov2.5_%', inplace = True)
dfOU
```

```python
dfOU.rename(columns = {'Ov2.5_%':'O_2.5_%'}, inplace=True)
#df5 = pd.merge(left=df4, right = dfOU, how='left')
```

```python
df5 = pd.merge(left=df4, right = dfOU, how='left')
```

```python
df5.head()
```

```python
# add home/away exp goals
df5['HG_Exp'] = round((df5['Exp']) * (df5['HW_%'] + 0.5 * df5['X_%']),2)
df5['AG_Exp'] = round((df5['Exp']) * (df5['AW_%'] + 0.5 * df5['X_%']),2)
df5['Sup Exp'] = df5['HG_Exp'] - df5['AG_Exp']
df5.head()
```

```python
df6 = df5.copy()
```

```python
# Calculate the league running average of 'team_a/b_offsides' column on a competition_id 
df6['h_lg_r_av'] = round(df6.groupby('competition_id')['team_a_offsides'].transform(lambda x: x.shift(1).expanding().mean()),2)
df6['a_lg_r_av'] = round(df6.groupby('competition_id')['team_b_offsides'].transform(lambda x: x.shift(1).expanding().mean()),2)
```

```python
# Calculate team running averages
def running_average(series):
    return series.expanding().mean().shift()

# Create running averages for when home team is in home_name column
df6['ht_h_r_av_f'] = round(df6.groupby(['competition_id', 'home_name'])['team_a_offsides'].apply(running_average), 2)
df6['ht_h_r_av_ag'] = round(df6.groupby(['competition_id', 'home_name'])['team_b_offsides'].apply(running_average), 2)

# Create running averages for when away team is in away_name column
df6['at_a_r_av_f'] = round(df6.groupby(['competition_id', 'away_name'])['team_b_offsides'].apply(running_average), 2)
df6['at_a_r_av_ag'] = round(df6.groupby(['competition_id', 'away_name'])['team_a_offsides'].apply(running_average), 2)

# Create running averages for when home team is in away_name column
df6['ht_a_r_av_f'] = round(df6.groupby(['competition_id', 'away_name'])['team_b_offsides'].apply(running_average), 2)
df6['ht_a_r_av_ag'] = round(df6.groupby(['competition_id', 'away_name'])['team_a_offsides'].apply(running_average), 2)

# Create running averages for when away team is in home_name column
df6['at_h_r_av_f'] = round(df6.groupby(['competition_id', 'home_name'])['team_a_offsides'].apply(running_average), 2)
df6['at_h_r_av_ag'] = round(df6.groupby(['competition_id', 'home_name'])['team_b_offsides'].apply(running_average), 2)

```

```python
df6.reset_index(drop=True,inplace=True)
df6.head()
```

```python
df6.shape
```

```python
df6.isnull().sum()
```

```python
df6.dropna(inplace=True)
```

```python
df6.shape
```

```python
# remove top 40 rows of each comp id - running averages will be skewed by lack of data
def remove_top_40_rows(group):
    return group.iloc[30:]

# Apply the function to each group within each competition_id
df_filtered = df6.groupby('competition_id').apply(remove_top_40_rows)

# Reset the index of the resulting DataFrame
df6 = df_filtered.reset_index(drop=True)
```
